```python
import pygame
import math

# Initialize Pygame
pygame.init()

# Constants
SCREEN_WIDTH = 1200
SCREEN_HEIGHT = 800
RING_RADIUS = 300
PLAYER_RADIUS = 305

PLAYER1_VELOCITY = 0.05
PLAYER2_VELOCITY = 0.05
PLAYER1_ARC_ANGLE = math.pi / 9  # 90 degrees in radians
PLAYER2_ARC_ANGLE = math.pi / 9  # 90 degrees in radians

# Colors
WHITE = (255, 255, 255)
BLACK = (0,0,0)
LGRAY = (190,200,210)
WGRAY = (160,160,160)
RED = (255, 0, 0)
GREEN = (0,255,0)
BLUE = (0,0,255)



# Create the game window
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
pygame.display.set_caption("Ring Game")

# Player position and angle (starts at the top of the ring)
player1_angle = -math.pi / 2  # -90 degrees in radians
player1_x = SCREEN_WIDTH // 2 + RING_RADIUS * math.cos(player1_angle)
player1_y = SCREEN_HEIGHT // 2 + RING_RADIUS * math.sin(player1_angle)
player2_angle = -math.pi / 2  # -90 degrees in radians
player2_x = SCREEN_WIDTH // 2 + RING_RADIUS * math.cos(player2_angle)
player2_y = SCREEN_HEIGHT // 2 + RING_RADIUS * math.sin(player2_angle)

# Main game loop
running = True
clock = pygame.time.Clock()

while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    keys = pygame.key.get_pressed()
    # Move the player on the ring
    if keys[pygame.K_a]:
        player1_angle -= PLAYER1_VELOCITY  # Adjust this value to control the movement speed
    if keys[pygame.K_d]:
        player1_angle += PLAYER1_VELOCITY  # Adjust this value to control the movement speed

    if keys[pygame.K_LEFT]:
        player2_angle -= PLAYER2_VELOCITY  # Adjust this value to control the movement speed
    if keys[pygame.K_RIGHT]:
        player2_angle += PLAYER2_VELOCITY  # Adjust this value to control the movement speed

    # Update player position based on the angle
    player1_x = SCREEN_WIDTH / 2
    player1_y = SCREEN_HEIGHT / 2
    player2_x = SCREEN_WIDTH / 2
    player2_y = SCREEN_HEIGHT / 2

    # Clear the screen
    screen.fill(LGRAY)

    # Draw the ring
    pygame.draw.circle(screen, WGRAY, (SCREEN_WIDTH / 2, SCREEN_HEIGHT / 2), RING_RADIUS)

    # Draw the player (arc)
    pygame.draw.arc(screen, BLUE, (player1_x - PLAYER_RADIUS, player1_y - PLAYER_RADIUS, PLAYER_RADIUS * 2, PLAYER_RADIUS * 2), player1_angle - PLAYER1_ARC_ANGLE / 2, player1_angle + PLAYER1_ARC_ANGLE / 2, width=5)
    pygame.draw.arc(screen, BLUE, (player1_x - PLAYER_RADIUS, player1_y - PLAYER_RADIUS, PLAYER_RADIUS * 2, PLAYER_RADIUS * 2), player1_angle - PLAYER1_ARC_ANGLE / 2, player1_angle + PLAYER1_ARC_ANGLE / 2, width=5)
    pygame.draw.arc(screen, RED, (player2_x - PLAYER_RADIUS, player2_y - PLAYER_RADIUS, PLAYER_RADIUS * 2, PLAYER_RADIUS * 2), player2_angle - PLAYER2_ARC_ANGLE / 2, player2_angle + PLAYER2_ARC_ANGLE / 2, width=5)
    pygame.draw.arc(screen, RED, (player2_x - PLAYER_RADIUS, player2_y - PLAYER_RADIUS, PLAYER_RADIUS * 2, PLAYER_RADIUS * 2), player2_angle - PLAYER2_ARC_ANGLE / 2, player2_angle + PLAYER2_ARC_ANGLE / 2, width=5)

    # Update the display
    pygame.display.flip()

    clock.tick(60)

# Quit Pygame
pygame.quit()
```
